```python
import numpy as np
import pandas as pd

#to identify will the user purchase from the store or not
```

```python
data=pd.read_csv('/content/Social_Network_Ads.csv') # importing file
```

```python
data.head() # checking the data header
```

```python
#will drop the non relavant colomns.
df=data.drop(['User ID'],axis=1)
```

```python
df.head() #Checking the status for Drop Command
```

```python
df['Gender']=df['Gender'].apply({'Male':1,'Female':0}.get) # converting text in to values
```

```python
df.head() # checking data header
```

```python
x=df[['Gender','Age','EstimatedSalary']] #independent
y=df['Purchased'] #dependent
```

```python
from sklearn.model_selection import train_test_split #importing function from sklearn
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8) # distributing the weightage in to train & test size
```

```python
from sklearn.neighbors import KNeighborsClassifier # importing Supervised algorithm from sklearn
```

```python
knn=KNeighborsClassifier(n_neighbors=5) # assigning sample number to start with algorithm
```

```python
knn.fit(x_train,y_train) # function usedto train the KNN in the training data set
```

```python
y_pred=knn.predict(x_test) # making prediction on testing data set
```

```python
predictions=knn.predict(x_test)# making prediction on testing data set
```

```python
print(predictions) # printing the prediction
```

```python
knn.score(x,y) # checking the accuracy score of above algorithm
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score # importing Confusion matrix from Sklearn module
```

```python
cm=confusion_matrix(y_test,y_pred) # applying confusion matrix
```

```python
ac=accuracy_score(y_test,y_pred) #Checking score or accuracy %
```

```python
print(cm)
```

```python
print(ac)
```
